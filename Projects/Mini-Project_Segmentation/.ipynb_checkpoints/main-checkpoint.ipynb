 # Find the top 10 Goalkeeper in Fifa 2019, just with GK skills

## Import libraries and Dataset

```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import squarify
from sklearn.preprocessing import StandardScaler
from yellowbrick.features.pca import PCADecomposition
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('fifa2019.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

# Data Cleaning and Transformation

## Drop columns with links and Jersey Number

```python
df = df.drop(columns=['Photo','Flag','Club Logo','Jersey Number'])
```

## Look for missing values

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
X = df._get_numeric_data()
X[X.isnull().any(axis=1)].head()
```

```python
X = X.dropna()
df = df.loc[(df['ID'].isin(X['ID'])),:]
```

# Filter just goalkeeper with AgglomerativeClustering

```python
X = df._get_numeric_data()
X = X.drop(columns=['Unnamed: 0', 'ID', 'Age', 'Overall', 'Potential', 'Special','International Reputation'])
hierarchical = AgglomerativeClustering(n_clusters=2, affinity='correlation', linkage='complete')
hierarchical.fit(X)
df['labels'] = hierarchical.labels_
df['labels'].unique()
```

```python
# look for position in label == 0
values = df.query('labels == 0')['Position'].value_counts()
squarify.plot(values, label=values.index,pad=True, alpha=.6 )
```

```python
# look for position in label == 1
values = df.query('labels == 1')['Position'].value_counts()
squarify.plot(values, label=values.index,pad=True, alpha=.6 )
```

```python
df_gk = df.query('labels == 1')
df_gk
```

# PCA Decompose with gk_skills

```python
df_gk_skills = df_gk[['GKDiving','GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']]
```

```python
X_std = StandardScaler().fit_transform(df_gk_skills)
pca = PCA(2)
X_t = pca.fit_transform(X_std)
pca.explained_variance_ratio_.sum()
```

```python
visualizer = PCADecomposition(scale=False, proj_features=True)
visualizer.fit_transform(pd.DataFrame(X_std, columns=df_gk_skills.columns))
visualizer.show()
```

# Group the best with Kmeans

```python
cluster = KMeans(n_clusters=4,random_state=42).fit(X_std)
labels= cluster.labels_
sns.scatterplot(X_t[:, 0], X_t[:, 1], hue=labels,s=50, palette='viridis', legend='full')
```

```python
df_gk['labels'] = labels
df_gk_good = df_gk.query('labels == 0')
```

```python
df_gk_good
```

# Create a rating to order a filter GK

```python
X=df_gk_good[['GKDiving','GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']]
X_std = StandardScaler().fit_transform(X)
pca = PCA(1)
X_t = pca.fit_transform(X_std)
pca.explained_variance_ratio_.sum()
```

```python
df_gk_good['rating'] = X_t
```

# Result / Compare with Overall

```python
df_gk_good=df_gk_good.sort_values(by='rating', ascending=False)
df_gk_good=df_gk_good.iloc[0:10,:]
df_gk_good[['Name','Age','Club','Overall','rating']]
```
