```python
import cv2
import numpy as np
from IPython.display import Image

img1 = cv2.imread('./cards/blue_0.png')
img2 = cv2.imread('./cards/blue_1.png')
vis = np.concatenate((img1, img2), axis=1)
cv2.imwrite('out1.png', vis)
Image(filename='out1.png')
```

```python
import numpy as np
import cv2
# Load the image
# Horizontally concatenate the 2 images
img3 = cv2.hconcat([img1, img2])
cv2.copyMakeBorder()
# Display the concatenated image
cv2.imwrite('out2.png', img3)
Image(filename= 'out2.png')
```
